```python
%matplotlib inline
from matplotlib import pyplot as plt
from scipy.stats import beta, norm
from numpy import linspace
from IPython.core.pylabtools import figsize
figsize(16, 8)
plt.style.use('ggplot')
```

```python
priorA = 1.0
priorB = 1.0
```

```python
viewsA = 6037
viewsB = 6040

clicksA = 3254
clicksB = 3576
```

```python
posteriorA = beta(clicksA + priorA, viewsA + priorB)
posteriorB = beta(clicksB + priorA, viewsA + priorB)
```

```python
print("Difference in means of posterior distributions A and B\n",
      posteriorB.mean() - posteriorA.mean())
```

```python
x = linspace(0, 1, 2000)
plt.plot(x, posteriorA.pdf(x), label='posterior of A')
plt.plot(x, posteriorB.pdf(x), label='posterior of B')
plt.xlabel('Value')
plt.ylabel('Density')
plt.title("Posterior distributions of the conversion rates between $A$ and $B$")
plt.xlim(0.32, 0.40)
plt.legend();
```

```python
nSamp = 1000000

samplesPosteriorA = posteriorA.rvs(nSamp)
samplesPosteriorB = posteriorB.rvs(nSamp)
```

```python
print("Probablity B convertions are superior to A\n",
      (samplesPosteriorB > samplesPosteriorA).mean())
```
