```python
%matplotlib inline
from matplotlib import pyplot as plt
from numpy.random import beta
plt.style.use('ggplot')
```

```python
nSamp = 1000000
```

```python
priorA = 1.0
priorB = 1.0
```

```python
viewsA = 6037
viewsB = 6040

clicksA = 3254
clicksB = 3576
```

```python
samplesA = beta(clicksA + priorA, viewsA + priorB, nSamp)
samplesB = beta(clicksB + priorA, viewsA + priorB, nSamp)
```

```python
plt.hist(samplesA, histtype='stepfilled', label='posterior of A', bins = 1000)
plt.hist(samplesB, histtype='stepfilled', label='posterior of B', bins = 1000)
plt.xlabel('Value')
plt.ylabel('Density')
plt.title("Sampled posterior distributions of the conversion rates of $A$ and $B$")
plt.legend();
```

```python
print("Probablity B convertions are superior to A\n",
      (samplesB > samplesA).mean())
```
